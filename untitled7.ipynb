```python
# =========================
# STEP 1 - IMPORT LIBRARIES
# =========================

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestClassifier

from sklearn.cluster import KMeans

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score

import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# =========================
# STEP 2 - UPLOAD DATASET
# =========================

from google.colab import files

uploaded = files.upload()
```

```python
# =========================
# STEP 3 - LOAD DATASET
# =========================

df = pd.read_csv('fitness_dataset_100.csv') # Changed filename to use an existing CSV file

print("First 5 Rows:")
print(df.head())
```

```python
# =========================
# STEP 4 - DATASET INFO
# =========================

print("\nDataset Information:")
print(df.info())
```

```python
# =========================
# STEP 5 - CHECK NULL VALUES
# =========================

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# =========================
# STEP 6 - REMOVE DUPLICATES
# =========================

df = df.drop_duplicates()

print("\nDataset Shape After Removing Duplicates:")
print(df.shape)
```

```python
# =========================
# STEP 7 - HANDLE MISSING VALUES
# =========================

# Numerical Columns
df['BMI'] = df['BMI'].fillna(df['BMI'].mean())

df['Steps'] = df['Steps'].fillna(df['Steps'].mean())

df['Sleep'] = df['Sleep'].fillna(df['Sleep'].mean())

# Categorical Columns
df['Goal'] = df['Goal'].fillna(df['Goal'].mode()[0])

df['Health'] = df['Health'].fillna(df['Health'].mode()[0])

df['Activity'] = df['Activity'].fillna(df['Activity'].mode()[0])
```

```python
# =========================
# STEP 8 - CLEAN TEXT DATA
# =========================

df['Goal'] = df['Goal'].str.strip().str.lower()

df['Health'] = df['Health'].str.strip().str.lower()

df['Activity'] = df['Activity'].str.strip().str.lower()
```

```python
# =========================
# STEP 9 - REMOVE OUTLIERS
# =========================

Q1 = df['BMI'].quantile(0.25)

Q3 = df['BMI'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR

upper_limit = Q3 + 1.5 * IQR

df = df[
    (df['BMI'] >= lower_limit) &
    (df['BMI'] <= upper_limit)
]

print("\nDataset Shape After Outlier Removal:")
print(df.shape)
```

```python
# =========================
# STEP 10 - CHECK CLASS BALANCE
# =========================

print("\nActivity Counts:")
print(df['Activity'].value_counts())
```

```python
# =========================
# STEP 11 - LABEL ENCODING
# =========================

le_goal = LabelEncoder()

le_health = LabelEncoder()

le_activity = LabelEncoder()

df['Goal'] = le_goal.fit_transform(df['Goal'])

df['Health'] = le_health.fit_transform(df['Health'])

df['Activity'] = le_activity.fit_transform(df['Activity'])
```

```python
# =========================
# STEP 12 - FEATURE SELECTION
# =========================

X = df[['Age', 'BMI', 'Steps', 'Sleep', 'Goal', 'Health']]

y = df['Activity']
```

```python
# =========================
# STEP 13 - FEATURE SCALING
# =========================

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

```

```python
# =========================
# STEP 14 - TRAIN TEST SPLIT
# =========================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
# =========================
# STEP 15 - TRAIN MODEL
# =========================

model = RandomForestClassifier(
    n_estimators=300,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    class_weight='balanced',
    random_state=42
)
model.fit(X_train, y_train)
```

```python
# =========================
# STEP 16 - PREDICTIONS
# =========================

predictions = model.predict(X_test)
```

```python

# =========================
# STEP 17 - MODEL ACCURACY
# =========================

accuracy = accuracy_score(y_test, predictions)

print("\nModel Accuracy:")
print(accuracy)
```

```python
# =========================
# STEP 18 - CLASSIFICATION REPORT
# =========================

print("\nClassification Report:")
print(classification_report(y_test, predictions))
```

```python
# =========================
# STEP 19 - CROSS VALIDATION
# =========================

scores = cross_val_score(
    model,
    X_scaled,
    y,
    cv=5
)

print("\nCross Validation Accuracy:")
print(scores.mean())
```

```python
# =========================
# STEP 20 - K-MEANS CLUSTERING
# =========================

kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

clusters = kmeans.fit_predict(X_scaled)

df['Cluster'] = clusters
```

```python
# =========================
# STEP 21 - VISUALIZATION
# =========================

plt.figure(figsize=(10,6))

sns.scatterplot(
    x=df['BMI'],
    y=df['Steps'],
    hue=df['Cluster'],
    palette='viridis'
)

plt.title("Fitness User Clusters")

plt.xlabel("BMI")

plt.ylabel("Daily Steps")

plt.show()
```

```python
# =========================
# STEP 22 - RECOMMENDATION FUNCTION
# =========================

def recommend(age, bmi, steps, sleep, goal, health):

    # Convert input text to lowercase
    goal = goal.lower().strip()

    health = health.lower().strip()

    # Encode goal and health
    goal_encoded = le_goal.transform([goal])[0]

    health_encoded = le_health.transform([health])[0]

    # Create dataframe
    input_data = pd.DataFrame([[
        age,
        bmi,
        steps,
        sleep,
        goal_encoded,
        health_encoded
    ]], columns=[
        'Age',
        'BMI',
        'Steps',
        'Sleep',
        'Goal',
        'Health'
    ])

    # Scale input
    input_scaled = scaler.transform(input_data)

    # Predict activity
    prediction = model.predict(input_scaled)

    # Decode prediction
    activity = le_activity.inverse_transform(prediction)

    return activity[0]
```

```python
# =========================
# STEP 23 - TEST THE SYSTEM
# =========================

result = recommend(
    age=25,
    bmi=23,
    steps=8500,
    sleep=7,
    goal='flexibility',
    health='asthma' # Changed 'none' to 'asthma' as it's a known label
)

print("\nRecommended Fitness Activity:")

print(result)
```

```python
# Collect user input interactively
my_age_input = int(input("Enter your age (e.g., 25): "))
my_bmi_input = float(input("Enter your BMI (e.g., 23.5): "))
my_steps_input = int(input("Enter your daily steps (e.g., 8500): "))
my_sleep_input = int(input("Enter your hours of sleep (e.g., 7): "))
my_goal_input = input("Enter your fitness goal (e.g., flexibility, weight loss): ")
my_health_input = input("Enter your health condition (e.g., none, diabetes): ")

# Get your recommendation using the interactive inputs
interactive_recommendation = recommend(
    age=my_age_input,
    bmi=my_bmi_input,
    steps=my_steps_input,
    sleep=my_sleep_input,
    goal=my_goal_input,
    health=my_health_input
)

print(f"\nBased on your interactive inputs, the recommended fitness activity is: {interactive_recommendation.capitalize()}")
```
